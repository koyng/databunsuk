```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv')
data=csv.reader(f)
next(data)
result=[]

for row in data :
    if row[-3] !='':
        if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14' :
            result.append(float(row[-1]))
plt.plot(result, 'r')
plt.show()
```
